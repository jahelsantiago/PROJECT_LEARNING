### Datos por categoria

```python
nom = ["gatos","perros","conejos"]
data = [100,200,50]
plt.figure(figsize = (20,5)) #ajusta el tamaño [ejex, ejey]

plt.subplot(131) #crea un subplot
plt.bar(nom,data)

plt.subplot(132)
plt.scatter(nom,data)

plt.subplot(133)
plt.plot(nom,data)



plt.show()
```

## Crear sub plots

```python
a = np.linspace(0,20,10)
def f(x):
    return np.exp(x)

plt.figure(figsize = (10,7))

plt.subplot(211) # con esto se actualizan crean nuevos subplot
plt.plot(a, f(a))
plt.xlabel("x")
plt.ylabel("Y")

plt.subplot(212)
plt.plot(a,np.cos(a))

```
